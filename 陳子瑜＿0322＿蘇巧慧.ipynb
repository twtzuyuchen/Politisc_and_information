輸入立委臉書粉專內容及選舉政見資料

```python
import pandas as pd
fb = pd.read_csv('nysu_10902_2019_research.csv')
politics = pd.read_csv('9th_legislator_promise.csv')
```

以新北市立委蘇巧慧為例，依照月份來統計數值資料的變化

```python
fb.head(1)
```

```python
su = fb[fb.page_name == "蘇巧慧"]
su.head(1)
```

```python
su[su.created_time_taipei == max(su.created_time_taipei)]
```

```python
su[su.created_time_taipei == min(su.created_time_taipei)]
```

```python
su['new_date'] = pd.to_datetime(su['created_time_taipei']).dt.date
su['month_year'] = pd.to_datetime(su['new_date']).dt.to_period('M')
su.head(1)
```

```python
results = su.groupby('month_year').sum()
results
```

```python
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
results['newhead'].map(str)
```

```python
import matplotlib.pyplot as plt

labels = results['newhead'].map(str)
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```

第十屆立委選舉日，民國109年（2020年）1月11號舉行，十一月為何聲量下跌？先以高聲量之十月分析。

```python
October = su[su.month_year == "2019-10"]
```

```python
results_october = October.groupby('new_date').sum()
results_october
```

```python
results_october.index.name = 'newhead'
results_october.reset_index(inplace=True)
results_october
```

```python
labels = [str(i) for i in range(1,32)]

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.plot(labels, results_october['like_count'])
ax.plot(labels, results_october['share_count'])
ax.plot(labels, results_october['comment_count'])

ax.legend()

plt.show()
```

輸入立委臉書粉專內容及選舉政見資料

```python
import pandas as pd
fb = pd.read_csv('nysu_10902_2019_research.csv')
politics = pd.read_csv('9th_legislator_promise.csv')
```

以新北市立委黃國昌為例，依照月份來統計數值資料的變化

```python
fb.head(1)
```

```python
huang = fb[fb.page_name == "黃國昌"]
huang.head(1)
```

```python
huang[huang.created_time_taipei == max(huang.created_time_taipei)]
```

```python
huang[huang.created_time_taipei == min(huang.created_time_taipei)]
```

```python
huang['new_date'] = pd.to_datetime(huang['created_time_taipei']).dt.date
huang['month_year'] = pd.to_datetime(huang['new_date']).dt.to_period('M')
huang.head(1)
```

```python
results = huang.groupby('month_year').sum()
results
```

```python
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
results['newhead'].map(str)
```

```python
import matplotlib.pyplot as plt

labels = results['newhead'].map(str)
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```

第十屆立委選舉日，民國109年（2020年）1月11號舉行，相較於選前兩個月，六月讚數高之原因爲何？而二月讚數最低，原因為何？

```python
June = huang[huang.month_year == "2019-06"]
```

```python
results_june = June.groupby('new_date').sum()
results_june
```

```python
results_june.index.name = 'newhead'
results_june.reset_index(inplace=True)
results_june
```

```python
import matplotlib.pyplot as plt

labels = [str(i) for i in range(1,28)]

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.plot(labels, results_june['like_count'])
ax.plot(labels, results_june['share_count'])
ax.plot(labels, results_june['comment_count'])

ax.legend()

plt.show()
```

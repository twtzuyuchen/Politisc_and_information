```python
yang = fb[fb.page_name == "楊曜Yangyao"]
yang['new_date'] = pd.to_datetime(yang['created_time_taipei']).dt.date
yang['post_hour'] = pd.to_datetime(yang['created_time_taipei']).dt.hour 
yang['month_year'] = pd.to_datetime(yang['new_date']).dt.to_period('M')
yang.head(1)
```

```python
post_time = yang[['post_id','post_hour']].groupby('post_hour').count()
post_time.index.name = 'post_hour'
post_time.reset_index(inplace = True)
post_time
```

```python
yang['new_date'] = pd.to_datetime(yang['created_time_taipei']).dt.date
yang['month_year'] = pd.to_datetime(yang['new_date']).dt.to_period('M')
yang.head(1)
```

```python
results = yang.groupby('month_year').sum()
results
```

```python
results.index.name = 'newhead'
results.reset_index(inplace=True)
results
```

```python
results['newhead'].map(str)
```

```python
import matplotlib.pyplot as plt

labels = results['newhead'].map(str)
width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

ax.bar(labels, results['like_count'], width, label='like_count')
ax.bar(labels, results['share_count'], width, label='share_count', bottom=results['like_count'])
ax.bar(labels, results['comment_count'], width, label='comment_count', bottom=results['share_count']+results['like_count'])

ax.legend()

plt.show()
```
